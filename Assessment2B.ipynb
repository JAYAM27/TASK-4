###### Department wise codekata performence (pie chart)

```python
dep = df['Department'].value_counts()
Department = ['CS', 'ECE', 'EEE']
vis = ['r', 'm', 'c']
myexplode = [0.1, 0.0, 0.0]
plt.pie(dep, labels = Department, colors = vis, shadow = True, explode = myexplode, autopct='%1.1f%%', startangle = 90, rotatelabels = True, )
plt.legend(title = 'Departments')
plt.show
```

###### Department wise toppers (horizantal bar graph or any visual representations of your choice)

```python
dep = df['Department'].value_counts()
fig, ax = plt.subplots()
Department = [ 'CS','ECE','EEE']
bar_labels = ['yellow', 'blue', 'orange']
vis = ['y', 'm', 'g']


ax.bar(Department, dep, label=bar_labels, color= vis)
ax.set_xlabel('Name of the Department')
ax.set_ylabel('Topper_score')
ax.set_title('Department wise toppers ')
ax.legend(title='Topper color')
plt.show()
```
